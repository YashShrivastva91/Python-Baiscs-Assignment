```python
#Q5. solution of Q5
#Example of Implicit Type Casting
x = 5       # int
y = 2.0     # float

result = x + y  # x is implicitly converted to float
print(result)   
print(type(result)) 

#Example of Explicit  Type Casting
#1.Converting String to Integer:
s = "123"
x = int(s)
print(x)        
print(type(x))  

#2.Converting Float to Integer:
f = 3.14
x = int(f)
print(x)        # Outputs: 3
print(type(x))

#3.Converting Integer to String:
n = 100
s = str(n)
print(s)        # Outputs: "100"
print(type(s))  # Outputs: <class 'str'>

#4Converting List to Tuple:
lst = [1, 2, 3]
t = tuple(lst)
print(t)        # Outputs: (1, 2, 3)
print(type(t))  # Outputs: <class 'tuple'>

#5.Converting Tuple to List:
t = (1, 2, 3)
lst = list(t)
print(lst)        # Outputs: [1, 2, 3]
print(type(lst))  # Outputs: <class 'list'>

#6.Converting List to Set:
lst = [1, 2, 2, 3]
s = set(lst)
print(s)        # Outputs: {1, 2, 3}
print(type(s))  # Outputs: <class 'set'>
```

```python
#Q3. example of ..Mutable Objects
#List
my_list = [1, 2, 3]
my_list[0] = 10
print(my_list)  # Outputs: [10, 2, 3]

#Dictionaries
my_dict = {'a': 1, 'b': 2}
my_dict['a'] = 10
print(my_dict)  # Outputs: {'a': 10, 'b': 2}

#sets
my_set = {1, 2, 3}
my_set.add(4)
print(my_set)  # Outputs: {1, 2, 3, 4}

## examples of Immutable Objects
#Strings
my_string = "hello"
new_string = my_string.replace('h', 'y')
print(my_string)  # Outputs: "hello"
print(new_string)  # Outputs: "yello"

#Tuples
my_tuple = (1, 2, 3)
new_tuple = my_tuple + (4,)
print(my_tuple)  # Outputs: (1, 2, 3)
print(new_tuple)  # Outputs: (1, 2, 3, 4)

##Integers
my_int = 5
new_int = my_int + 3  # Adding 3 to my_int
print(new_int)  # Outputs: 8
```

```python
#Q.6 Example
# 'if' Statement:
x = 10

if x > 5:
    print("x is greater than 5")
    
# 'if-else' Statement:
x = 3

if x > 5:
    print("x is greater than 5")
else:
    print("x is 5 or less")
    
# 'if-elif-else' Statement:
x = 7

if x > 10:
    print("x is greater than 10")
elif x > 5:
    print("x is greater than 5 but 10 or less")
else:
    print("x is 5 or less")
    
# 'Nested if' Statements:
x = 15
y = 10

if x > 10:
    if y > 5:
        print("x is greater than 10 and y is greater than 5")
    else:
        print("x is greater than 10 but y is 5 or less")
else:
    print("x is 10 or less")
    
#Ternary Conditional Operator:
x = 8
result = "x is even" if x % 2 == 0 else "x is odd"
print(result)

```

```python
#Q7. for loop example 
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
    
# Iterating over a list:
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
    
#Using range() function:    
for i in range(5):
    print(i)
    
#Iterating over a dictionary:   
student_grades = {"Yash": 90, "Suraj": 85, "Amar": 88}
for student, grade in student_grades.items():
    print(f"{student}: {grade}")
    
#Basic 'while loop':
count = 0
while count < 5:
    print(count)
    count += 1
    
#Stopping condition:
user_input = ""
while user_input != "exit":
    user_input = input("Type 'exit' to stop: ")
    print("You typed:", user_input)
    
```
